```python
snpeff= open("/Users/cmdb/qbb2020-answers/QuantLabHomework/Assignment2/snpeff.vcf", "r")
snpeff = snpeff.readlines()
snpdict = {}
gen_qual= []

for line in snpeff: 
    if "#" in line: 
        continue
    else:
        line = line.rsplit()
        form = line[9:-1]
        for sublist in form: 
            sublist = sublist.split(":")
            if sublist[1] == '.':
                continue
            else:
                gen_qual.append(float(sublist[1]))
```

```python
allele_freq = []
for line in snpeff:
    if "#" in line: 
        continue
    #line = line.rsplit()[7].split(';')
    else: 
        find = line.find('AF=')
        freqline = line[find+3:find+10]
        #freqline= float(freqline.split(';')[0])
        freqline= freqline.split(';')[0].split(',')
        print(freqline)
        allele_freq.append(freqline)

allele_freq = sum(allele_freq, []) # Flatten nested lists
allele_freq = [float(i) for i in allele_freq] # Convert to floats
#print(allele_freq)
        
        
        
        
        #for i in freqline: 
            #if "," in i: 
                #i = i.split(',')
                #print(i[0])
                
                
    
    
        
```

```python
pred_effects = []
for line in snpeff:
    if "#" in line: 
        continue
    #line = line.rsplit()[7].split(';')
    else: 
        find = line.find('ANN=')
        pred = line[find:-1]
        #freqline= float(freqline.split(';')[0])
        pred = pred.split('\t')[0].split('|')
        #print(pred)
        effects = pred[1].split('&') 
        pred_effects.append(effects)
        #print(pred_effects)
        #print(effects)
#print(pred_effects)            
         # Flatten nested lists
pred_effects = sum(pred_effects, [])
        
#allele_freq = sum(allele_freq, []) # Flatten nested lists
#allele_freq = [float(i) for i in allele_freq] # Convert to floats
#print(allele_freq)

pred_count = {x:pred_effects.count(x) for x in pred_effects}
#print(pred_count)     
```

```python
sort_pred_count = {k: pred_count[k] for k in sorted(pred_count, key=pred_count.get, reverse=True)}

```
